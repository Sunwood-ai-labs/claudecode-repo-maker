## 1. 基本的なPythonコード

簡単な計算を実行してみます。

## 2. リストとループ

Pythonのリストとループを使った例です。

```python
# リストの作成
numbers = [1, 2, 3, 4, 5]

# 各要素の二乗を計算
for num in numbers:
    print(num ** 2)
```

## 3. 関数の定義

カスタム関数を定義して使用します。

```python
def factorial(n):
    """階乗を計算する関数"""
    if n <= 1:
        return 1
    return n * factorial(n - 1)

# 5の階乗を計算
factorial(5)
```

## 4. データ構造

辞書を使ったデータ管理の例です。

```python
# 辞書の作成
person = {
    "name": "Alice",
    "age": 30,
    "city": "Tokyo"
}

# 辞書の内容を表示
for key, value in person.items():
    print(f"{key.capitalize()}: {value}")
```
